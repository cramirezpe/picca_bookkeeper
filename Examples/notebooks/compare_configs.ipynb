```python
def replace_bookkeeper_string(text, bookkeeper):
    text = re.sub(
        str(bookkeeper.paths.fits_path),
        'fits-path',
        text,
    )
    
    text = re.sub(
        str(bookkeeper.paths.correlations_path),
        'correlations-path',
        text,
    )
    
    text = re.sub(
        str(bookkeeper.paths.run_path),
        'run-path',
        text,
    )
    
    text = re.sub(
        str(bookkeeper.paths.catalog),
        'catalog',
        text,
    )
    
    text = re.sub(
        str(bookkeeper.paths.catalog_dla),
        'catalog-dla',
        text,
    )
    text = re.sub(
        str(bookkeeper.paths.catalog_bal),
        'catalog-bal',
        text,
    )
    text = re.sub(
        str(bookkeeper.paths.catalog_tracer),
        'catalog-tracer',
        text,
    )
    
    text = re.sub(
        str(r".*ini files =.*"),
        'ini files',
        text,
    )
            
    text = re.sub(
        str(r".*zeff =.*"),
        'ini files',
        text,
    )
    
    text = re.sub(
        "\(",
        "",
        text
    )
    
    text = re.sub(
        "\)",
        "",
        text
    )
    
    text = re.sub(
        '--',
        "'\n'",
        text,
    )
    
    
    return text
    
```

```python
for bookkeeper_config in basedir.glob('**/*.yaml'):
    bookkeeper = Bookkeeper(bookkeeper_config)
    
    print(bookkeeper_config.name[:-4], "\n-------------")
    
    
    if bookkeeper.paths.catalog != baseline.paths.catalog:
        print('Different QSO catalog:\n\t', bookkeeper.paths.catalog)
        
    if bookkeeper.paths.catalog_dla != baseline.paths.catalog_dla:
        print('Different DLA catalog:\n\t', bookkeeper.paths.catalog_dla)
    
    if bookkeeper.paths.catalog_bal != baseline.paths.catalog_bal:
        print('Different BAL catalog:\n\t', bookkeeper.paths.catalog_bal)
        
    if bookkeeper.paths.catalog_tracer not in (baseline.paths.catalog_tracer, baseline.paths.catalog):
        print('Different tracer catalog:\n\t', bookkeeper.paths.catalog_tracer)
        
    
    
    ini_files = list((bookkeeper.paths.run_path / "configs").glob("*.ini"))
    script_files = list((bookkeeper.paths.correlations_path / "scripts").glob("*.sh"))
    fit_files = list((bookkeeper.paths.fits_path / "configs").glob("*.ini"))
    
    ini_files_base = [baseline.paths.run_path / "configs" / x.name for x in ini_files]
    script_files_base = [baseline.paths.correlations_path / "scripts" / x.name for x in script_files]
    fit_files_base = [baseline.paths.fits_path / "configs" / x.name for x in fit_files]
    
    config_files = ini_files + script_files + fit_files 
    baseline_files = ini_files_base + script_files_base + fit_files_base
    
    for config_file, baseline_file in zip(config_files, baseline_files):
        print('\t', config_file.name)
            
        
        baseline_text = replace_bookkeeper_string(
            baseline_file.read_text(),
            baseline,
        )
        
        run_text = replace_bookkeeper_string(
            config_file.read_text(),
            bookkeeper,
        )
        
        
        #!diff <(echo "{baseline_text}") <(echo "{run_text}") --color
        !diff <(echo "{baseline_text}") <(echo "{run_text}") --color --palette='ad=1;3;38;5;135:de=1;3;38;5;9'
        
        #'ad=1;3;38;5;154:de=1;3;38;5;9'
        

    print('')   
    print('################################\n'*3)
```
